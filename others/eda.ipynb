```python
import databento as db
import pandas as pd

# === 1. Load DBN File ===
path = "CLX5_mbo.dbn"
store = db.DBNStore.from_file(path)
df = store.to_df()

print("✅ Loaded", len(df), "rows")
print("Columns:", list(df.columns))
print("\nUnique symbols:", df["symbol"].unique())
print("Time range:", df.index.min(), "→", df.index.max())

# === 2. Basic Info ===
print("\n--- Basic Statistics ---")
# Try to handle datetime safely
numeric_cols = df.select_dtypes(include=["number"]).columns
print(df[numeric_cols].describe())

# === 3. Check Sorting & Continuity ===
print("\n--- Order Checks ---")
print("Timestamps sorted:", df.index.is_monotonic_increasing)
print("Sequence sorted:", df["sequence"].is_monotonic_increasing)
print("Unique order_id count:", df["order_id"].nunique())

# Check if multiple instruments exist
print("Instrument IDs:", df["instrument_id"].unique())

# === 4. Action Distribution ===
if "action" in df.columns:
    print("\n--- Action Type Distribution ---")
    print(df["action"].value_counts(dropna=False))

# === 5. Side Distribution ===
if "side" in df.columns:
    print("\n--- Side (Buy/Sell) ---")
    print(df["side"].value_counts(dropna=False))

# === 6. Time Interval Gaps ===
df["ts_recv_diff"] = df.index.to_series().diff().dt.total_seconds() * 1e3  # in ms
print("\n--- Timing ---")
print("Average inter-message gap (ms):", df["ts_recv_diff"].mean())
print("Max gap (ms):", df["ts_recv_diff"].max())

# === 7. Duplicates or Missing Values ===
print("\n--- Data Integrity ---")
print("Duplicated order_ids:", df["order_id"].duplicated().sum())
print("Missing values per column:\n", df.isna().sum())
```
